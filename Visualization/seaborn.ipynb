```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# List of Dataset 
sns.get_dataset_names()
```

```python
# Load mpg dataset
mpg = sns.load_dataset('mpg')
```

```python
mpg
```

```python
sns.set_theme()
```

```python
sns.set_style('whitegrid')
sns.set_style('ticks')
sns.set_style('darkgrid')
```

```python
sns.set_context('poster',font_scale=1.2) # larger
sns.set_context('talk',font_scale=1.2) # large
sns.set_context('paper',font_scale=1.2) # normal
```

```python
# correlation Data
mpg.corr(numeric_only=True)
```

```python
sns.heatmap(mpg.corr(numeric_only=True));
```

```python
plt.figure(figsize=(20,10)) # Width, Height
sns.scatterplot(data=mpg,x='weight',y='mpg', hue='origin', palette='mako'); # Use hue that doesn't have many variances
sns.despine(left=True)
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(data=mpg,x='model_year',y='mpg'); # Use hue that doesn't have many variances
```

```python
plt.figure(figsize=(20,10))
sns.pointplot(data=mpg,x='model_year',y='mpg'); # Use hue that doesn't have many variances
```

```python
sns.kdeplot(data=mpg,x='weight',y='mpg'); 
```

```python
# Presenting wise, easy to visualize
sns.kdeplot(data=mpg,x='weight',y='mpg', fill=True); 
```

```python
sns.histplot(data=mpg,x='weight',y='mpg', fill=True); 
```

```python
sns.histplot(data=mpg,x='mpg'); 
```

```python
sns.histplot(mpg['mpg']); # Same as sns.displot(mpg['mpg']); 
```

```python
sns.histplot(mpg['mpg'], kde=True, bins=20); # bins: number of boxes
```

```python
sns.kdeplot(mpg['mpg']); 
```

```python
sns.ecdfplot(data=mpg,x='mpg');
```

```python
sns.countplot(data=mpg,x='origin',palette='mako');
```

```python
sns.barplot(data=mpg,x='origin',y='mpg');
```

```python
sns.boxplot(data=mpg,x='origin',y='mpg',palette='mako');
# The dots on top are the outliers, middle blue box means 50% and 25% each top and bottom.
```

```python
# Little bit more sophistigated box plot:
sns.boxenplot(data=mpg,x='origin',y='mpg',palette='mako');
```

```python
# Little bit more sophistigated box plot:
sns.violinplot(data=mpg,x='origin',y='mpg',palette='mako');
```

```python
# Little bit more sophistigated box plot:
sns.stripplot(data=mpg,x='origin',y='mpg',palette='mako');
```

```python
# mixture of scatter and histogram
sns.jointplot(data=mpg,x='weight',y='mpg');
```

```python
# mixture of scatter and histogram
sns.jointplot(data=mpg,x='weight',y='mpg',kind='reg');
```

```python
# mixture of scatter and histogram
sns.jointplot(data=mpg,x='weight',y='mpg',kind='hex');
```

```python
# mixture of scatter and histogram
sns.jointplot(data=mpg,x='weight',y='mpg',kind='kde');
```

```python
# mixture of scatter and histogram
sns.jointplot(data=mpg,x='weight',y='mpg',kind='kde',fill=True);
```
