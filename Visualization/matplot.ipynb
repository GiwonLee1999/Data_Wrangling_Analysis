```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
x = [1,2,3,4,5]
y = [0,2,4,6,13]

# Resize Graph
plt.figure(figsize=(5,3), dpi=150)

# Argument Notation
plt.plot(x,y , label='2x', color='r',linewidth=2, marker='^', markersize=10, markeredgecolor='b', linestyle='--')

# Line Number 2
x2 = np.arange(0,4,0.3)
y2 = np.arange(0,10,0.5)

# Graph.2 Continuation after certain point
plt.plot(x2[:8],x2[:8]**2, label='x^2')
plt.plot(x2[7:],x2[7:]**2, linestyle='--')

# Graph Title, labels
plt.title('Graph', fontdict={'fontname':'Comic Sans MS','fontsize':20});

plt.xlabel('x label');
plt.ylabel('y label');

# Graph indexes
plt.xticks([0,1,2,3])
plt.yticks([2,3,4,5,12])

# Graph Legend
plt.legend()

# Save Graph in the same directory
plt.savefig('mygraph.png',dpi=250)

plt.show()
```

## Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1,2,3]

bars = plt.bar(labels,values)

patterns = ['/','O','*']
for bar in bars:
    bar.set_hatch(patterns.pop())

# bars[0].set_hatch('/')
# bars[1].set_hatch('O')
# bars[2].set_hatch('*')

plt.figure(figsize=(6,4),dpi=200)

plt.show()
```
