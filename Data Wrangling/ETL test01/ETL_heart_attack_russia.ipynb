```python
data = pd.read_csv('heart_attack_russia_youth_vs_adult.csv')
len(data)
```

```python
data.isna() # false means no missing data
```

```python
missing_data = data.isna().sum()
missing_data
```

```python
# Based off of this, we can see that there are no missing values in the dataset except for the Alcohol consumption
data['Alcohol_Consumption']
data['Alcohol_Consumption'].fillna('None', inplace=True)
data['Alcohol_Consumption'].isna().sum()
# Now we have no missing values in the Alcohol Consumption column
```

```python
# Sample data is created
sample_data = data.dropna() # drops the rows with missing values
len(sample_data)
```

```python
sample_data.head()
```

```python
missing_count = sample_data.isna().sum()
missing_value_percentage = missing_count / len(sample_data) * 100
# missing_value_percentage
region_group = sample_data.groupby(['Region']).size()
gender_group = sample_data.groupby(['Gender']).size()
age_group = sample_data.groupby(['Age']).size()
# Here we can see the number of people in each age group and 28 year olds are the most heart attacks.
# age_group.sort_values(ascending=False)
```

##### Given the data, I am able to analyze the relationship between various factors, including **age, gender, cholesterol levels, body mass index (BMI), smoking habits, stress levels, and sleep duration**.

Certainly, each of the columns can contribute to a heart attack, but they were excluded for being of lesser significance.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# sns.scatterplot(x='Age', y='Sleep_Hours', data=sample_data)
plt.figure(figsize=(10, 10))
sns.boxplot(x='Age', y='Sleep_Hours', data=sample_data)

plt.show()
```

```python
# This graph shows the average Cholesterol levels for each Age/BMI group

avg_cholesterol = sample_data.groupby(['Age'])['Cholesterol'].mean()
sns.scatterplot(x='Age', y='BMI', hue=avg_cholesterol, data=sample_data, palette="coolwarm", s=100)
plt.title('Age vs BMI colored by Average Cholesterol Levels')
plt.show()
```

```python
# Tyring out this cool plot, it shows the relationship between the Age, BMI, and Cholesterol levels

sns.pairplot(sample_data[['Age', 'BMI', 'Cholesterol']], hue='Cholesterol', palette="coolwarm")
plt.show()
```

```python
# This average_cholesterol variable is used to calculate the average Cholesterol levels for each age group
average_cholesterol = sample_data.groupby('Age')['Cholesterol'].mean().reset_index()

bins = [10, 30, 40, 50, 60]  
labels = ['10-30', '31-40', '41-50', '51-60']
sample_data['age_group'] = pd.cut(sample_data['Age'], bins=bins, labels=labels)

sns.lineplot(x='Age', y='Cholesterol', data=average_cholesterol, marker='o')
plt.title('Average Cholesterol by Age')
plt.xlabel('Age')
plt.ylabel('Average Cholesterol')
plt.show()
```
